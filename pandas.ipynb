```python
def grade(x):
  if x>=90:
   return 'a'
  elif x>=80:
   return 'b'
  elif x>=70:
   return'c'
  elif x>=60:
   return'd'
  elif x>=50:
   return'e'
  else:
   return'f'

df['grade'] = df['exam_score'].apply(grade)
df
```

```python
import pandas as pd

df1 = pd.DataFrame({
     'id': [1,2,3,4],
     'name' : ['rahul','kabir','rayyan','syed'],
     'age' :  [25,30,35,40]
}
)


df2 = pd.DataFrame({
    'id' : [3,4,5,6],
    'name':['kabir','rayyan','syed','vamshi'],
    'age':[25,30,35,40]
})


df3 = pd.merge(df1,df2,on = 'id',how = 'inner')
df3 = pd.merge(df1,df2,on = 'id',how = 'left')
df3 = pd.merge(df1,df2,on = 'id',how = 'right')
df3 = pd.merge(df1,df2,on = 'id',how = 'outer')
print(df3)
```

```python
import pandas as pd
df1 = pd.DataFrame({
     'id': [1,2,3,4],
     'name' : ['rahul','rhea','ravi','raja'],
     'age' :  [25,30,35,40]
}
)


df2 = pd.DataFrame({
    'id' : [3,4,5,6],
    'name':['kabir','rayyan','syed','vamshi'],
    'age':[25,30,35,40]
})


df= pd.concat([df1, df2])

display(df)
```
