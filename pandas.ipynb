```python
def grade(x):
  if x>=90:
   return 'a'
  elif x>=80:
   return 'b'
  elif x>=70:
   return'c'
  elif x>=60:
   return'd'
  elif x>=50:
   return'e'
  else:
   return'f'

df['grade'] = df['exam_score'].apply(grade)
df
```

```python
df = pd.read_csv('student_data.csv')
df.head()
```

```python
sns.lineplot(data=df,x='study_hours',y='exam_score')
```

```python
def grade(x):
  if x>=90:
   return 'a'
  elif x>=80:
   return 'b'
  elif x>=70:
   return'c'
  elif x>=60:
   return'd'
  elif x>=50:
   return'e'
  else:
   return'f'

df['grade'] = df['exam_score'].apply(grade)
df
```

```python
sns.barplot(data=df,x='grade',y='exam_score',estimator='mean')
plt.title('performance of students')
plt.xlabel('grade')
plt.ylabel('exam_score')

plt.tight_layout()
plt.show()
```
