```python
def grade(x):
  if x>=90:
   return 'a'
  elif x>=80:
   return 'b'
  elif x>=70:
   return'c'
  elif x>=60:
   return'd'
  elif x>=50:
   return'e'
  else:
   return'f'

df['grade'] = df['exam_score'].apply(grade)
df
```
